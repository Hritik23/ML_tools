# Activation Functions 

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def print_act(st, xmin=-10, xmax=10, style='plot'):
    x=(np.arange(xmin,xmax,0.05))
    y=[eval(st+"("+str(c)+")") for c in x]
    
    plt.figure(figsize=(7,5))
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(st)
    plt.grid()
    if style=='plot':
        plt.plot(x,y)
    else:
        plt.scatter(x,y,s=1)
    plt.show()
```

### Identity

```python
def identity_function(x):
    return x
print_act('identity_function')
```

### Binary step

```python
def binary_step(x):
    if x>=0:
        return 1
    else:
        return 0
print_act("binary_step", style='scatter')
```

### Piece Wise Linear

```python
def piece_wise_linear(x, xmin=-3, xmax=5):
    if x>=xmax:
        return 1
    elif x<=xmin:
        return 0
    else:
        m=1/(xmax-xmin)
        b=1-m*xmax
        return m*x+b
print_act("piece_wise_linear")
```

### Bipolar

```python
def bipolar(x):
    if x>0:
        return 1
    else:
        return -1


print_act('bipolar', style='scatter')
```

### Sigmoid

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
print_act('sigmoid')
```

### Complementary log-log

```python
def comp_log_log(x):
    return 1-math.exp(-math.exp(x))

print_act('comp_log_log')
```

### Bipolar Sigmoid

```python
def bipolar_sigmoid(x):
    return ( 1-math.exp(-x) )/( 1+math.exp(-x) )
print_act("bipolar_sigmoid")
```

### tanh

```python
def tanh(x):
    return math.tanh(x)
print_act('tanh')
```

### Hard tanh

```python
def hard_tanh(x):
    return max(-1,min(1,x))

print_act('hard_tanh')
```

### LeCun's tanh

```python
def lecun_s_tanh(x):
    return 1.7159*math.tanh(2*x/3)

print_act("lecun_s_tanh")
```

### Rectified Linear Unit (ReLU)

```python
def relu(x):
    if x>=0:
        return x
    else:
        return 0
print_act('relu')
```

### Leaky Rectified Linear Unit (Leaky ReLU)

```python
def leaky_relu(x):
    if x>=0:
        return x
    else:
        return 0.1*x
print_act('leaky_relu')
```

### Parametric Rectified Linear Unit (Parametric ReLU)

```python
def parametric_relu(x,alpha=.4):
    if x>=0:
        return x
    else:
        return alpha*x
print_act("parametric_relu")
```

### Smooth Rectified Linear Unit (Smooth ReLU) / Softplus / SmoothMax

```python
def softplus(x):
    return math.log(1+math.exp(x))
print_act('softplus')
```

### Exponential Linear Unit (ELU)

```python
def exponential_linear_unit(x, a=0.5): # here, a is hyperparameter
    if x>=0:
        return x
    else:
        return a*(math.exp(x)-1)
print_act('exponential_linear_unit')
```

### Scaled Exponential Linear Unit (SELU)

```python
def selu(x, scale=1.0507, alpha=1.6732): #here, scale and alpha are pre-defined constants
    if x >=0:
        return scale*x
    else:
        return scale*alpha*(math.exp(x)-1)
    
print_act('selu')
```

### Thresholded Rectified Linear Unit (Thresholded ReLU)

```python
def thresholded_relu(x, theta=2):
    if x > theta:
        return x
    else:
        return 0
print_act("thresholded_relu", style='scatter')
```

### Modifications to Relu (Unofficial)

```python
def relu_sin(x):
    return relu(x)+math.sin(x)
print_act('relu_sin')
```

```python
def relu_cos(x):
    return relu(x)+math.cos(x)
print_act('relu_cos')
```

### Softsign

```python
def soft_sign(x):
    return x/(1+abs(x))
print_act("soft_sign")
```

### Exponential

```python
def exponential(x):
    return math.exp(x)
print_act('exponential', -3, 2)
```

### Absolute

```python
def absolute(x):
    return abs(x)
print_act("absolute")
```

### Swish

```python
def swish(x):
    return x*sigmoid(x)

print_act('swish')
```
